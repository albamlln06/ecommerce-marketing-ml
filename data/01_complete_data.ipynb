```python
import sys
import os
import src.utils.translator as translator
import src.utils.data_load as data_load
import pandas as pd

BATCH_SIZE = 100
OUTPUT_PATH = '../data/data_files/reviews_translated.csv'

datasets = data_load.load_datasets()
df_reviews = datasets["reviews"]

# Si ya existe un archivo parcial, retomar desde donde se quedó
if os.path.exists(OUTPUT_PATH):
    df_done = pd.read_csv(OUTPUT_PATH)
    start_idx = len(df_done)
    print(f"Retomando desde fila {start_idx} de {len(df_reviews)}")
else:
    df_done = None
    start_idx = 0

total = len(df_reviews)
all_batches = []

for batch_start in range(start_idx, total, BATCH_SIZE):
    batch_end = min(batch_start + BATCH_SIZE, total)
    batch = df_reviews.iloc[batch_start:batch_end].copy()

    print(f"Traduciendo filas {batch_start}–{batch_end - 1} / {total}...")

    try:
        batch_translated = translator.translate_reviews(batch, "review_comment_message")
    except Exception as e:
        print(f"Error en batch {batch_start}–{batch_end - 1}: {e}")
        break

    all_batches.append(batch_translated)

    # Guardar progreso acumulado tras cada batch
    df_so_far = pd.concat(
        ([df_done] if df_done is not None else []) + all_batches,
        ignore_index=True
    )
    df_so_far.to_csv(OUTPUT_PATH, index=False)
    print(f"  ✓ Guardado hasta fila {batch_end - 1}")

print("Traducción completada.")
```

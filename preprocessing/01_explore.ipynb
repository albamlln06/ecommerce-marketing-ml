```python
import matplotlib.pyplot as plt
import numpy as np

# Pedidos únicos por cliente (sin duplicados por join)
orders_per_customer = (
    df_complete.drop_duplicates(subset='order_id')
    .groupby('customer_unique_id')['order_id']
    .nunique()
)

counts = orders_per_customer.value_counts().sort_index()

# Bucketing (a partir de 5 agrupar)
buckets = {}
for n, c in counts.items():
    key = str(int(n)) if n <= 4 else '5+'
    buckets[key] = buckets.get(key, 0) + c

labels = list(buckets.keys())
values = list(buckets.values())
total  = sum(values)

# Graficar
fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.bar(labels, values, color='#378ADD', edgecolor='white', linewidth=0.8)

for bar in bars:
    h = bar.get_height()
    pct = h / total * 100
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        h + total * 0.003,
        f'{int(h):,} ({pct:.1f}%)',
        ha='center', va='bottom', fontsize=10
    )

ax.set_xlabel('Nº de pedidos realizados')
ax.set_ylabel('Nº de clientes')
ax.set_title('Distribución de pedidos por cliente')
ax.set_ylim(0, max(values) * 1.12)
ax.spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.show()

# Resumen
print(f"Total clientes únicos:          {len(orders_per_customer):,}")
print(f"Clientes con 1 solo pedido:     {(orders_per_customer == 1).sum():,} ({(orders_per_customer == 1).mean()*100:.1f}%)")
print(f"Clientes con más de 1 pedido:   {(orders_per_customer > 1).sum():,} ({(orders_per_customer > 1).mean()*100:.1f}%)")
print(f"Máximo pedidos por cliente:     {orders_per_customer.max()}")
print(f"Media pedidos por cliente:      {orders_per_customer.mean():.2f}")
```

```python
import matplotlib.pyplot as plt

# Calcular
order_size = (
    df_complete.drop_duplicates(subset=['order_id', 'product_id_x'])
    .groupby('order_id')['product_id_x']
    .nunique()
    .rename('order_size')
)

df_clean = df_complete.drop_duplicates(subset=['order_id', 'product_id_x']).merge(order_size, on='order_id')

product_context = df_clean.groupby('product_id_x').agg(
    en_pedido_multi = ('order_size', lambda x: (x > 1).any()),
    en_pedido_solo  = ('order_size', lambda x: (x == 1).any()),
).reset_index()

def classify(row):
    if row['en_pedido_multi'] and row['en_pedido_solo']:
        return 'En ambos tipos'
    elif row['en_pedido_multi']:
        return 'Solo en pedidos multi-producto'
    else:
        return 'Solo en pedidos de 1 producto'

product_context['categoria'] = product_context.apply(classify, axis=1)
counts = product_context['categoria'].value_counts()

# Graficar
labels = counts.index.tolist()
values = counts.values.tolist()
total  = sum(values)
colors = ['#378ADD', '#7F77DD', '#1D9E75']

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.bar(labels, values, color=colors, edgecolor='white', linewidth=0.8)

for bar in bars:
    h = bar.get_height()
    pct = h / total * 100
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        h + total * 0.005,
        f'{int(h):,} ({pct:.1f}%)',
        ha='center', va='bottom', fontsize=10
    )

ax.set_ylabel('Nº de productos')
ax.set_title('Productos según contexto de pedido en que aparecen')
ax.set_ylim(0, max(values) * 1.12)
ax.spines[['top', 'right']].set_visible(False)
plt.xticks(wrap=True)
plt.tight_layout()
plt.show()
```

```python


# Frecuencia de productos y categorías
df_products = df_products.merge(df_categories, on='product_category_name', how='left')
df_products['category'] = (
    df_products['product_category_name_english']
    .fillna(df_products['product_category_name'])
    .fillna('unknown')
)

df_items_enriched = df_order_items.merge(
    df_products[['product_id', 'category']], on='product_id', how='left'
)

product_freq   = df_items_enriched['product_id'].value_counts()
category_freq  = df_items_enriched['category'].value_counts()

thresholds = range(1, 21)

product_counts  = [( product_freq >= t).sum() for t in thresholds]
category_counts = [(category_freq >= t).sum() for t in thresholds]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Cobertura por umbral mínimo de pedidos', fontsize=14, fontweight='bold')

for ax, counts, label, color, total in zip(
    axes,
    [product_counts, category_counts],
    ['Productos', 'Categorías'],
    ['#4C72B0', '#DD8452'],
    [product_freq.shape[0], category_freq.shape[0]]
):
    ax.bar(list(thresholds), counts, color=color, alpha=0.85, edgecolor='white')

    ax.axvline(x=5, color='red', linestyle='--', linewidth=1.5, label='Umbral = 5')
    val_at_5 = counts[4]
    ax.axhline(y=val_at_5, color='red', linestyle=':', linewidth=1, alpha=0.6)

    ax.annotate(
        f'{val_at_5:,}\n({val_at_5/total:.1%} del total)',
        xy=(5, val_at_5),
        xytext=(8, val_at_5 * 0.95),
        fontsize=10,
        color='red',
        fontweight='bold',
        arrowprops=dict(arrowstyle='->', color='red', lw=1.5)
    )

    ax.set_xlabel('Mínimo de pedidos', fontsize=11)
    ax.set_ylabel(f'Número de {label.lower()}', fontsize=11)
    ax.set_title(f'{label} con al menos N pedidos', fontsize=12)
    ax.set_xticks(list(thresholds))
    ax.legend(fontsize=10)
    ax.grid(axis='y', alpha=0.3)
    ax.set_xlim(0.5, 20.5)

    ax.text(0.98, 0.97, f'Total: {total:,}',
            transform=ax.transAxes, ha='right', va='top',
            fontsize=9, color='gray')

plt.tight_layout()
plt.savefig('coverage_threshold.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\n{'Umbral':>8} {'Productos':>12} {'% Productos':>13} "
      f"{'Categorías':>12} {'% Categorías':>13}")
print("─" * 62)
for t, p, c in zip(thresholds, product_counts, category_counts):
    marker = " ◄" if t == 5 else ""
    print(f"{t:>8} {p:>12,} {p/product_freq.shape[0]:>12.1%}  "
          f"{c:>12,} {c/category_freq.shape[0]:>12.1%}{marker}")
```
